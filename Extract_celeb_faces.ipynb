```python
import cv2
import os
from PIL import Image
import numpy as np
path = "E://Paper_publication//review_2//Celeb-df//images//Celeb-real-images//"
dirs = os.listdir(path)
i=1
for item in dirs:
    if os.path.isfile(path+item):
        img = cv2.imread(path+item) #Path of an image
        gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
#         faces = faceCascade.detectMultiScale(img,1.1,4)
        faces=faceCascade.detectMultiScale(gray_img,scaleFactor=1.18,minNeighbors=5,flags=cv2.CASCADE_SCALE_IMAGE)
        name = 'E:\\Paper_publication\\review_2\\datasets_preprocessing\\train_images\\Celeb-real-face\\' + item 
#         print(i)
        for (x, y, w, h) in faces:
            FaceImg = img[y:y+h,x:x+w]
            # To save an image on disk
            print ('Creating...' + name)
            pil_image = Image.fromarray(FaceImg)
            faceimg=pil_image.resize((224, 224))
            FaceImg=np.array(faceimg)
            cv2.imwrite(name,FaceImg)
        i+=1
```
